```python
#TABLE TwO
CustomerHistory = df[['RowNumber',
                      'CustomerId',
                      'Tenure',
                      'HasCrCard',
                      'IsActiveMember',
                      'Balance'
                     ]]
```

```python
CustomerHistory
```

```python
CustomerHistory.replace({'IsActiveMember':{1:'Active', 0:'Inactive'}}, inplace = True)
```

```python
CustomerHistory.replace({'Tenure':{0:'Semi annually',
                                  1:'Annually',
                                  2:'Two years',
                                  3:'Three years',
                                  4:'Four years',
                                  5:'Five years',
                                  6:'Six years',
                                  7:'Seven years',
                                  8:'Eight years',
                                  9:'Nine years',
                                  0:'Ten years'}}, inplace = True)
```

```python
CustomerHistory.replace({'HasCrCard':{0:'NO', 1:'YES'}}, inplace = True)
```

```python
CustomerHistory
```

```python
#TABLE THREE
ProductData= df[['RowNumber',
               'HasCrCard',
               'NumOfProducts',
                'CreditScore',
               'Balance']]
```
